```python
mixed_lst=[1, "hello", 3.14, True]

print(mixed_lst)
```

```python
#Accessing list elements
print(mixed_lst[1])  # Output: hello
print(mixed_lst[-1]) # Output: True
print(mixed_lst[0:2]) # Output: [1, 'hello']
print(mixed_lst[:3])  # Output: [1, 'hello', 3.14]
print(mixed_lst[::2]) # Output: [1, 3.14]   
```

```python
#Modifying list elements
mixed_lst[1] = "world"
print(mixed_lst)  # Output: [1, 'world', 3.14, True]
mixed_lst[2:4] = [2.71, False]
print(mixed_lst)  # Output: [1, 'world', 2.71, False]
```

```python
#List methods
mixed_lst.append("new item")
print(mixed_lst)  # Output: [1, 'world', 2.71, False, 'new item']
print(len(mixed_lst))  # Output: 5
print(mixed_lst.count(False))  # Output: 1
print(mixed_lst.index("world"))  # Output: 1
mixed_lst.insert(2, "inserted item")
print(mixed_lst)  # Output: [1, 'world', 'inserted item', 2.71, False, 'new item']
print(mixed_lst.pop())  # Output: 'new item'
```

```python
#removing items
print(mixed_lst)  # Output: [1, 'world', 'inserted item', 2.71, 'new item']
#mixed_lst.clear()
print(mixed_lst)  # Output: []

```

```python
mixed_lst.index("world")

mixed_lst.reverse()
print(mixed_lst)  # Output: []
```

```python
#iterating through a list
for item in mixed_lst:
    print(item)
# Output: (no output since the list is empty)

```

```python
#list comprehension
squared_lst = [x**2 for x in range(5)]
print(squared_lst)  # Output: [0, 1, 4, 9, 16]

cubed_even_lst = [x**3 for x in range(11) if x % 2 == 0]
print(cubed_even_lst)  # Output: [0, 8, 64, 216, 512, 1000]

quadratic_lst = [x**2 for x in range(10)]
print(quadratic_lst)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
lst1 = [x**2 for x in range(1,11)]
print(lst1)

lst2 = [x**3 for x in range(1,11) if x%2==0]
print(lst2)
```

```python
#nested lists
nested_lst = [[1, 2, 3], ['a', 'b', 'c'], [True, False]]
print(nested_lst[1][0])  # Output: 'a'
```
