```python
#tuple methods

tpl = (1, 2, 3, 2, 4, 2, 5)
count_2 = tpl.count(2)
print(count_2)  # Output: 3
index_4 = tpl.index(4)
print(index_4)  # Output: 4
mixed_tpl = (1, "hello", 3.14, True)
print(mixed_tpl)  # Output: (1, 'hello', 3.14, True)
print(mixed_tpl.index("hello"))  # Output: 1
print(mixed_tpl.count(2))  # Output: 0
print(mixed_tpl.index(3.14))  # Output: 2
#mixed_tpl.index("world")  # Raises ValueError: tuple.index(x): x not in tuple
print(mixed_tpl.count(True))  # Output: 1
print(mixed_tpl.index(1))  # Output: 0
print(mixed_tpl.count("hello"))  # Output: 1
#mixed_tpl.index(100)  # Raises ValueError: tuple.index(x): x not in tuple
print(mixed_tpl.count(3.14))  # Output: 1
print(mixed_tpl.index(True))  # Output: 3
print(mixed_tpl.count("world"))  # Output: 0
#mixed_tpl.index(2)  # Raises ValueError: tuple.index(x): x not in tuple
print(mixed_tpl.count(1))  # Output: 1
print(mixed_tpl.index(3.14))  # Output: 2
print(mixed_tpl.count(False))  # Output: 0
print(mixed_tpl.index("hello"))  # Output: 1

```

```python
#packing and unpacking
point = (10, 20)
x, y = point
print(x)  # Output: 10
print(y)  # Output: 20


#unpacking with asterisk
data = (1, 2, 3, 4, 5)
a, b, *rest,d = data
print(a)     # Output: 1
print(b)     # Output: 2
print(rest)  # Output: [3, 4, 5]
print(d)     # Output: 5
```

```python
#nested tuples
nested_tpl = (1, (2, 3), (4, (5, 6)))
print(nested_tpl[0])        # Output: 1 
print(nested_tpl[1])        # Output: (2, 3)
print(nested_tpl[2])        # Output: (4, (5, 6))
print(nested_tpl[2][1])     # Output: (5, 6)
print(nested_tpl[2][1][0])  # Output: 5

print(nested_tpl[1][0])     # Output: 2
print(nested_tpl[1][1])     # Output: 3
print(nested_tpl[2][0])     # Output: 4
print(nested_tpl[2][1][1])  # Output: 6
```
